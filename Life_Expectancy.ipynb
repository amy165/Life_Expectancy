```python
Life_Expectancy_clean_encoded=pd.get_dummies(Life_Expectancy_clean, columns=['Status'], dtype=int)
Life_Expectancy_clean_encoded.head()
```

```python
X=Life_Expectancy_clean_encoded.drop('Life expectancy ', axis=1)
y=Life_Expectancy_clean_encoded['Life expectancy ']
```

```python
X=np.array(X)
Y=np.array(y)
```

```python
Y=Y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
import xgboost as xgb
model=xgb.XGBRegressor(objective='reg:squarederror', learning_rate=0.1, max_depth=30, n_estimators=200)
model.fit(X_train, Y_train)
```

```python
result = model.score(X_test, Y_test)
print("Accuracy : {}".format(result))
```

```python
Y_predict = model.predict(X_test)
```

```python
Y_predict
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(Y_test, Y_predict)),'.3f'))
MSE = mean_squared_error(Y_test, Y_predict)
MAE = mean_absolute_error(Y_test, Y_predict)
r2 = r2_score(Y_test, Y_predict)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2)
```

```python
#el valor de R^2 es cercano a 1 por lo que se puede decir que es buen resultado
```
